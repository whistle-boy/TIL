## 3주차 과제
#### 문제 1

```python
N = int(input('3~18 사이 정수 N을 입력해주세요.'))


def die3(x, y, z):
    #     print('세 번째 주사위 : ', z)
    #     print(' = ', x, y, z)
    if (x + y + z) == N:
        print(x, y, z)


def die2(x, y):
    #     print('두 번째 주사위 : ', y)
    for z in range(1, 7):
        die3(x, y, z)


def die1(x):
    #     print('첫 번째 주사위 : ', x)
    for y in range(1, 7):
        die2(x, y)


def main():
    global N
    while N not in range(3, 18+1):
        N = int(input('3~18 사이 정수 N을 입력해주세요.'))
    for x in range(1, 7):
        die1(x)


main()
```

#### 문제 2

```python
def program2():
    N = int(input('양의 정수 N을 입력하세요. N:'))
    M = int(input('양의 정수 M을 입력하세요. M:'))
    if N >= M:
        high = N
        low = M
    else:
        high = M
        low = N
    quotient = high // low
    remainder = high - (low*quotient)
    print(quotient, remainder, sep='\n')


program2()
```

#### 문제 4

```python
def program4():
    N = int(input('10이상의 양의 정수 N을 입력하세요. N:'))
    while N >= 10:
        N = sum(int(digit) for digit in str(N))
    print(N)


program4()
```
