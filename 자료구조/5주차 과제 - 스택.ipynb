```python
class Stack:
    def __init__(self):  # 생성자
        self.ls = []    # top이 이제 클래스의 멤버 변수가 됨
        
    def isEmpty(self): return len(self.ls) == 0
    def size(self): return len(self.ls)
    def clear(self): self.ls = []
        
    def push(self, item):
        self.ls.append(item)
        
    def pop(self):
        if not self.isEmpty():
            return self.ls.pop(-1)
        
    def peek(self):
        if not self.isEmpty():
            return self.ls[-1]
```

```python
def checkBrackets(statement):
    N = 0
    stack = Stack()
    for ch in statement:             # 괄호 개수
        if ch in ('{', '[', '(', '}', ']', ')'):
            N += 1
    for ch in statement:             # 문자열의 각 문자에 대해
        if ch in ('{', '[', '('):    # in '{[('도 동일하게 동작함
            stack.push(ch)
        elif ch in ('}', ']', ')'):  # in'}])' 도 동일하게 동작함
            if stack.isEmpty():
                return print('Wrong_', N, sep='')        # 조건 2 위반(왼쪽 괄호 먼저)
            else:
                left = stack.pop()
                if (ch == '}' and left != '{') or \
                    (ch == ']' and left != '[') or \
                    (ch == ')' and left != '('):
                    return print('Wrong_', N, sep='')    # 조건 3 위반(괄호 사이 포함 관계)
                
    if not stack.isEmpty():           # False이면 조건 1 위반(괄호 개수 일치)
        return print('Wrong_', N, sep='')
    
    else:
        return print('OK_', N, sep='')
```

```python
statement = [
    '(3+40*(2+(30-7)*2133)',
    '3*{4+(2-792)/1} + [3*{4-2* (100 -7)}]',
    '301*{4+(2101-7)/1} + 9*{4-2* (10108-7)}}',
    '(3*{4001+(2-7)/1} + [3*{4-2* (1-7)}])'
            ]
for s in statement:
    checkBrackets(s)
```
