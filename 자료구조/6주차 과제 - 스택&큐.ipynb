## 문제 1-스택

```python
class Stack:
    def __init__(self):
        self.top = []
        
    def isEmpty(self): return len(self.top) == 0
    def size(self): return len(self.top)
    def clear(self): self.top = []
        
    def push(self, item):
        self.top.append(item)
        
    def pop(self):
        if not self.isEmpty():
            return self.top.pop(-1)
        else:
            print('Stack Empty')
        
    def peek(self):
        if not self.isEmpty():
            print(self.top[-1])
    
    def duplicate(self):
        data = self.top.pop(-1)
        for i in range(2):
            self.push(data)
            
    def upRotate(self, n):
        if n <= self.size():
            left = self.top[-n:-1]
            right = self.top[-1:]
            self.top[-n:-1] = right
            self.top[-1:] = left
            
    def downRotate(self, n):
        if n <= self.size():
            left = self.top[-n:-n+1]
            right = self.top[-n+1:]
            self.top[-n:-1] = right
            self.top[-1:] = left
            
    def print_stack(self):
        temp = ''.join(self.top[::-1])
        print(temp)
        
```

## 문제 2-큐

```python
MAX_QSIZE = int(input('q = ')) # 원형 큐의 크기
```

```python
class CircularQueue:
    def __init__(self):
        self.MAX_QSIZE = int(input('q = '))
        self.front = 0
        self.rear = 0
        self.items = [0] * MAX_QSIZE
    def isEmpty(self): return self.front == self.rear
    def isFull(self): return self.front == (self.rear+1)%MAX_QSIZE
    def enqueue(self, item):
        if not self.isFull():
            self.rear = (self.rear+1)%MAX_QSIZE
            self.items[self.rear] = item
        else:
            print('overflow', self.items)
            return
            
    def dequeue(self):
        if not self.isEmpty():
            self.front = (self.front+1)%MAX_QSIZE
            self.items[self.front] = 0
        else:
            print('underflow')
            return
                    
    def peek(self):
        if not self.isEmpty():
            return self.items[(self.front + 1)%MAX_QSIZE]
        
    def size(self):
        return (self.rear - self.front + MAX_QSIZE) % MAX_QSIZE
    
    def display(self):
        print(self.items)
        
def program_():
    n = int(input('n = ')) # 연산의 개수
    for i in range(n):
        temp = input().split()
        if temp[0] == 'I': que.enqueue(temp[1])
        elif temp[0] == 'D': que.dequeue()
        elif temp[0] == 'P': que.display()
```

```python
que = CircularQueue()

program_()
```
