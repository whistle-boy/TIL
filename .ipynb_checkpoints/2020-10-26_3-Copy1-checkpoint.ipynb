```python
fig, ax = plt.subplots(1,2, figsize=(16,8))

raw_data['운동을 할 충분한 시간이 없어서'].plot.pie(explode=[0,0.02],\
                                       ax=ax[0], autopct='%1.1f%%')
ax[0].set_title('운동을 할 충분한 시간이 없어서')
ax[0].set_ylabel('')

raw_data['함께 운동을 할 사람이 없어서'].plot.pie(explode=[0,0.02],\
                                      ax=ax[1], autopct='%1.1f%%')
ax[1].set_title('함께 운동을 할 사람이 없어서')
ax[1].set_ylabel('')

plt.show()
```

# 서울시 범죄 현황 분석

### 일단 데이터를 한 번 읽고 판단해 보자

```python
import numpy as np
import pandas as pd
```

```python
crime_raw_data = pd.read_csv('./data/2017_crime.csv', encoding='euc-kr')
crime_raw_data.head()
```

```python
# read_csv 명령으로 서울시 구별 범죄현황 데이터를 읽었다
```

```python
crime_raw_data = crime_raw_data[crime_raw_data['죄종'].notnull()]
crime_raw_data.info()
```

```python
# 혹시 모를 null에 대비
```

- 애초의 목적은?
    - 강남 3구가 안전한지 확인하는 것
- 현재 데이터는
    - 경찰서 이름, 죄종 등으로 그냥 나열된 데이터
- 우리는 구별로 정리된 데이터가 필요하다

### 데이터 정리의 마법 명령어 -  pivot_table

### pivot_table을 이용한 데이터 정리

- 이게 우리 원 데이터(raw data)였다
- 경찰서별로 잘 정리된 데이터가 필요하다
- 방법은?
    - pivot_table

```python
crime_station = crime_raw_data.pivot_table(
    crime_raw_data, index=['구분'], columns=['죄종', '발생검거'], aggfunc=[np.sum])

crime_station.head()
```

```python
# 그런데 이 복잡한 칼럼들은 뭐지?
```

```python
crime_station.columns
```

```python
crime_station.columns = crime_station.columns.droplevel([0,1])
crime_station.columns
```

```python
# 일단 제일 필요없는 상위 두 개의 칼럼 이름은 제거하자
```

```python
crime_station['강도', '검거']
```

```python
# 이 상태에서는 이렇게 접근할 수 있다. 강도의 검거 건수는?
```

```python
crime_station['살인', '발생']
```

```python
# 살인이 발생된 건수는?
```

```python
crime_station.head()
```

```python
# 그러나 정리가 되긴 했지만, 현재 2중 컬럼 이름은 불필요해 보인다
# 오히려 컬럼을 강도검거, 강도발생 등으로 한 줄로 보는 것이 더 편해 보임
```

```python
tmp = crime_station.columns.get_level_values(0) + \
    crime_station.columns.get_level_values(1)

tmp
```

```python
# 이렇게 두 개의 컬럼을 더해보면 된다
```

```python
crime_station.columns = tmp
crime_station.head()
```

```python
# 그리고 그 두 컬럼을 합친 하나의 컬럼을 선택하면 된다
```

```python
# 문제는 이 데이터들이 구별로 정리된 것이 아니라 경찰서별로 정리된 것이라는 것
# 그렇다면 경찰서 이름으로 경찰서가 위치해 있는 구 이름을 알아야 한다는 것
```

### 데이터를 채우자

```python
crime_station.index
```

```python
station_names = ['혜화', '남대문', '중부', '종암', '서부', '방배', '수서']
gu_names = ['종로구', '중구', '중구', '성북구', '은평구', '서초구', '강남구']
```

```python
tmp_gu = []
for idx, _ in crime_station.iterrows():
    if idx in station_names:
        tmp_gu.append(gu_names[station_names.index(idx)])
    else:
        tmp_gu.append(idx+'구')
```

```python
crime_station['구'] = tmp_gu
crime_station
```

```python
crime_gu = pd.pivot_table(crime_station, index='구', aggfunc=np.sum)

crime_gu.head()
```

```python
crime_gu['살인발생'].sort_values(ascending=False).head()
```

```python
crime_gu['폭력발생'].sort_values(ascending=False).head()
```

```python
crime_gu['강도발생'].sort_values(ascending=False).head()
```

```python
crime_gu['절도발생'].sort_values(ascending=False).head()
```

### Pandas 컬럼간 연산(나눗셈)

```python
crime_gu.head()
```

```python
# 만약, 검거율(검거건/발생건)을 만들고 싶다면?
```

```python
crime_gu['강도검거']/crime_gu['강도발생']
```

```python
# 이렇게. 해도 된다. 5대 범죄니까 5번하면 된다.
# 그러나...
# 두 개 이상의 컬럼일 때는 div
# 분자가 두 개 일 때, 동시에 나눠준다
# 그냥 나눗셈은 에러가 난다
```

```python
crime_gu[['강도검거', '살인검거']].div(crime_gu['강도발생'], axis=0)
```

```python
# 여러 컬럼을 한 컬럼으로 나눌 때는 이렇게 해도 된다
# 우리가 원하는 건 아니지만
```

```python
num= ['강간검거', '강도검거', '살인검거', '절도검거', '폭력검거']
den= ['강간발생', '강도발생', '살인발생', '절도발생', '폭력발생']

crime_gu[num].div(crime_gu[den].values)
```

```python
# 각각의 컬럼을 분모(den), 분자(num)로 나눠주는 것은 위의 방법
```

```python
target = ['강간검거율', '강도검거율', '살인검거율', '절도검거율', '폭력검거율']
num = ['강간검거', '강도검거', '살인검거', '절도검거', '폭력검거']
den= ['강간발생', '강도발생', '살인발생', '절도발생', '폭력발생']

crime_gu[target] = crime_gu[num].div(crime_gu[den].values)*100
crime_gu.head()
```

```python
# 여기에 만들어질 컬럼까지 지정해주면 OK
```

```python
crime_gu = crime_gu.drop(columns=num)
crime_gu.head()
```

```python
# 검거율을 사용하고, 검거 컬럼들(5개)은 삭제
```

```python
# 검거 시점의 문제로 100%가 넘는 검거율이 존재한다
# 그대로 둬도 되지만, 나중에 그래프에서의 일관성을 위해 손을 대자...
```

```python
crime_gu[target] > 100
```

```python
# 위와 같이 조건을 주면 100이상인 경우면 True를 받을 수 있다
```

```python
crime_gu[crime_gu[target] > 100] = 100
crime_gu
```

```python
# 그럴때 강제로 100으로 맞추게 한다
# 단, 이 코드는 일부 Pandas 버전에서는 동작하지 않는다
```

```python
crime_gu.rename(columns = {'강간발생':'강간',
                          '강도발생':'강도',
                          '살인발생':'살인',
                          '절도발생':'절도',
                          '폭력발생':'폭력'},
               inplace=True)
crime_gu.head()
```

```python
# 일부 컬럼의 이름을 변경한다
```

```python
# 5대 범죄의 수치적 차이가 극심하다
# 문제는 5대 범죄를 한 번에 시각화할때 살인같은 경우는 잘 들어나지 않는다
# 범죄간 경중을 따질 수 없지만, 각 컬럼을 정규화하여 최대값을 같게 만든다
```

```python
col = ['살인', '강도', '강간', '절도', '폭력']
crime_gu_norm = crime_gu[col] / crime_gu[col].max()
crime_gu_norm.head()
```

```python
# 모든 요소가 양수이므로 정규화할 컬럼을 정하고,
# 각 컬럼의 최대값을 각각의 컬럼에 나눠주면 된다
# 주의) 새로 DataFrame을 만들었다.(crime_gu_norm)
```

```python
col2 = ['강간검거율', '강도검거율', '살인검거율', '절도검거율', '폭력검거율']
crime_gu_norm[col2] = crime_gu[col2]
crime_gu_norm.head()
```

```python
# 그리고 나머지 컬럼을 또 가져와서 맞췄다
# 이제 발생건들은 최대값이 1
# 검거는 최대값이 100이다
```

```python
col = ['강간', '강도', '살인', '절도', '폭력']
crime_gu_norm['범죄'] = np.mean(crime_gu_norm[col], axis=1)
crime_gu_norm.head()
```

```python
# 범죄라는 컬럼으로 범죄 발생의 평균치를 저장
# 컬럼 방향 연산이 아니라 row 방향 연산이라 axis=1 옵션을 지정
```

```python
col = ['강간검거율', '강도검거율', '살인검거율', '절도검거율', '폭력검거율']
crime_gu_norm['검거'] = np.mean(crime_gu_norm[col], axis=1)
crime_gu_norm.head()
```
