- hello.py : app 객체를 생성, route 설정
- static : js, css, image파일 등을 저장
- templates : html 코드를 저장

```python
%%writefile hello/hello.py
from flask import *

app = Flask(__name__)

# app 객체 안에 route라고 하는 decorator 함수
@app.route('/')
def hello():
    return 'hello world'

@app.route('/user/<name>')
def user(name):
    return render_template('index.html', name=name)

@app.route('/api/data')
def api_data():
    data = {'alice':25, 'andy':35}
    return jsonify(data)

app.run(debug=True)
```
