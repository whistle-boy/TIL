```python
from bs4 import BeautifulSoup
```

```python
url = "http://corners.gmarket.co.kr/Bestsellers"
response = requests.get(url)
response
```

```python
response.text
```

```python
dom = BeautifulSoup(response.text, "html.parser")
```

```python
dom
```

```python
selector = "#gBestWrap > div > div:nth-child(5) > \
div:nth-child(3) > ul > li"
elements = dom.select(selector)
len(elements)
```

```python
element = elements[0]
```

```python
data = {
    "title": element.select_one(".itemname").text.strip(),
    "link": element.select_one(".itemname").get("href"),
    "o-price": element.select_one(".o-price").text.strip(),
    "s-price": element.select_one(".s-price").text.strip().split("\n")[0],
}

data
```

```python
datas = []
for element in elements:
    data = {
        "title": element.select_one(".itemname").text.strip(),
        "link": element.select_one(".itemname").get("href"),
        "o-price": element.select_one(".o-price").text.strip(),
        "s-price": element.select_one(".s-price").text.strip().split("\n")[0],
    }
    datas.append(data)
```

```python
def save_file(title, link):
    response = requests.get(link)
    with open("datas/{}.png".format(title), "wb") as f:
        f.write(response.content)
```

### 크롤링 vs 스크래핑

- 크롤링
    - 스스로 돌아다니면서 웹에서 정보를 얻는 행위
- 스크래핑
    - 웹페이지 하나에서 정보를 얻는 행위
- 파싱(리딩)
    - 어떤 정보를 얻는다
