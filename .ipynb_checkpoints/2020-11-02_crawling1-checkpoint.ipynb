### web crawling
- 1. requests : json : 동적페이지(url변경 없이 웹페이지 내용이 변경)
    - naver stock, zigbang
- 2. requests : html : 정적페이지(url변경으로 웹페이지 내용이 변경)
    - gmarket
- 3. selenium : web browser : 1, 2번 방법을 사용하지 못할 때 사용

#### 1. 네이버 주가 데이터 크롤링
- 코스피, 코스닥 데이터
    - 로그인 없이 크롤링 가능
- usd 환율 데이터
    - 로그인 해야 크롤링 가능
    - header에 cookie 값을 적용해서 request를 실행
- 그래프 그리기
- 정규화 : min max scale을 사용
- 상관계수

#### 크롤링 절차
- 1. 웹서비스 분석 : url을 알아내기
- 2. request, response 하기 : json 포맷의 str 데이터 받기
- 3. json(str) > dict > dataframe

```python
# response를 보고 json타입인지 html타입인지 확인
```

```python
# 1. 웹서비스 분석 : url을 알아내기
```

```python
# code : 'KOSPI', 'KOSDAQ'
code, page_size, page = 'KOSPI', 20, 1
url = f'https://m.stock.naver.com/api/json/sise/dailySiseIndexListJson.nhn?code={code}&pageSize={page_size}&page={page}'

print(url)
```

```python
import requests
```

```python
# 2. request, response 하기 : json 포맷의 str 데이터 받기
```

```python
response = requests.get(url) # get방식으로 동작하기 때문에 header를 보면 request method에 get인지 post인지 나와있음
response
```

```python
type(response.text) # response.content
```

```python
# 3. json(str) > dict > dataframe
```

```python
# dir로 method를 확인하기
dir(response)
```

```python
# json method가 str > dict
datas = response.json()
type(datas)
```

```python
datas
```

```python
datas = response.json()['result']['siseList']
type(datas)
```

```python
import pandas as pd
```

```python
kospi_df = pd.DataFrame(datas)
kospi_df.tail()
```

```python
# 함수로 만들기
```

```python
def naver_stock(code='KOSPI', page_size=20, page=1):
    url = f'https://m.stock.naver.com/api/json/sise/dailySiseIndexListJson.nhn?code={code}&pageSize={page_size}&page={page}'
    response = requests.get(url) # get방식으로 동작하기 때문에 header를 보면 request method에 get인지 post인지 나와있음
    return pd.DataFrame(response.json()['result']['siseList'])
```

```python
kospi_df = naver_stock(page_size=100)
print(len(kospi_df))
kospi_df.head(2)
```

```python
kosdaq_df = naver_stock(code='KOSDAQ', page_size=100)
print(len(kosdaq_df))
kosdaq_df.head(2)
```

#### 정규화 : normalization
- z = (x - min(x)) / (max(x) - min(x))

```python
!pip install sklearn
```

```python
from sklearn import preprocessing
```

```python
preprocessing.minmax_scale(kospi_df['ncv'])
```

```python
plt.figure(figsize=(20, 5)) # 가로20, 세로5
plt.plot(kospi_df['dt'], preprocessing.minmax_scale(kospi_df['ncv']), 'r*-', label='kospi')
plt.plot(kosdaq_df['dt'], preprocessing.minmax_scale(kosdaq_df['ncv']), 'bo:', label='kosdaq')
plt.plot(usd_df['dt'], preprocessing.minmax_scale(usd_df['nv']), 'g^--', label='usd')

# 라벨 데이터
plt.legend(loc=0)

# 데이터 갯수에서 10을 나눈 몫
step = len(kospi_df) // 10
# ::step으로 원하는 갯수만 나온다
plt.xticks(kospi_df[::step]['dt'])

plt.show()
```
