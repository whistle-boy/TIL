#### Server & Client Architecture
- 웹서비스 기본 구조
- Client
    - request
- Server
    - response
- 전반적인 구조
    - client >> browser >> url & request >> server >> wep application & db >> response.. html, excel 등.. >> browser가 해석 >> client
- 크롤링
    - browser를 쓰지 않고 request/response하여 python을 써서 df로 만드는 것

#### url
- uniform resource locator
- protocol
    - http, https, ...
    - encoding과 비슷한 것
    - https 서비스는 id/password를 암호화 >> server가 푼다
        - 요즘엔 chrome 검색어 https가 아닌 사이트는 상단에 잘 뜨지 않음
- domain
    - 어떤 서버로 요청할지 알려주는 것
    - ip로 바꿔준다
- 예제
    - http://news.naver.com:80/main/read.nhn?model=LSD&mid=shm&sid1=105&oid=001&aid=0009847211#da_727145
        - https:// - protocol
        - news - Sub Domain
        - naver.com - Domain(ip)
        - 80 - port(문)
        - /main/ - path(디렉토리)
        - read.nhn - page(파일, '?' 앞)
        - ?mode=LSD&mid=shm&sid1=105&oid=001&aid=0009847211 - query(보여줄 정보, '?' 뒤)
        - #da_727145 - fragment(특정 위치에 관한 데이터)

#### get & post
- get

    - 길이 제한이 있음
- post
    - body에 데이터가 포함 >> 데이터가 숨겨짐

#### cookie & session & cache
- cookie
    - client browser에 저장하는 문자열 데이터
    - 도메인 별로 따로 저장
    - 로그인 정보, 내가 봤던 상품 정보, 팝업 다시보지 않음
- session
    - 연결정보에 관한 객체
    - browser별로 session 정보가 다르다
- cache
    - server의 메모리에 저장하여 빠르게 데이터를 가져오는 목적의 저장소

#### http status code
- 상태 코드
    - 2xx - success
    - 3xx - redirection
        - 요청 >> 응답을 받는다 >> html을 browser가 캐시에 저장을 한다 >> 다시 요청을 한다 >> 다시 요청하면 네트워크 비용이 드니까 저장되어 있는 것을 화면에 보여준다 >> 대신 내용이 수정되면 과거 데이터를 가져오기 때문에 내용이 안 바뀔 수 있다
    - 4xx - 요청 에러
        - 404 에러 - 도메인 에러.. 도메인에 해당하는 서버가 없다
    - 5xx - 서버 에러

        - 저걸로 데이터 수집해서 요청이 올 때 적당한 것을 잘 보여주는 것
    - 구글 봇
        - 사이트를 다 들어간다(요청) >> 데이터를 가져와서 db에 쌓음(사이트에 대한 정보가 db에 차곡차곡)

### web crawling
- 1. requests : json : 동적페이지(URL 변경 없이 데이터를 수정 및 추가)
- 2. requests : html : 정적페이지(URL 변경으로 데이터를 수정 및 추가)
- 3. selenium : web browser : 1, 2번 방법을 사용하지 못할때 사용
    - 서버 쪽으로 데이터 요청하는 것이 아니라 browser에서 데이터를 가져오는 것
    - 서버입장에서는 클라이언트가 데이터를 가져가면 트래픽 발생하여 서버입장에서 비용이 발생.. 그래서 데이터를 가져가지 못하게 서버 쪽에서는 막음.. 막는 방법이 보통 browser가 일반적인 browser가 아닐 때.. like 파이썬!
    - selenium은 브라우저를 띄워서 데이터를 가져오기 때문에 거의 실패 없지만, 시간이 훨씬 많이 걸린다
- 모바일 홈페이지가 보통 더 심플하기 때문에, 모바일 홈페이지가 있는지 확인

#### KOSDAQ 크롤링

```python
# code부분만 kosdaq으로 바꿔주게되면 kosdaq을 가져온다..

# 함수로 만들기
def get_stock_data(code, page_size=20, page=1):
    url = f"https://m.stock.naver.com/api/json/sise/dailySiseIndexListJson.nhn?\
code={code}&pageSize={page_size}&page={page}"
    response = requests.get(url)
    datas = response.json()["result"]["siseList"]
    kosdaq_df = pd.DataFrame(datas)
    return kosdaq_df
```

```python
kosdaq_df = get_stock_data('KOSDAQ', 100)
kosdaq_df.tail(2)
```

```python
kospi_df = get_stock_data('KOSPI', 100) # 100개 데이터 가져오기
kospi_df.tail(2)
```

#### 정규화(normalization)
- `z = (x - min(x)) / (max(x) - min(x))`
- 0~1 사이 값

```python
from sklearn import preprocessing
```

```python
preprocessing.minmax_scale(merge_df["ncv_kospi"])
```

```python
# 각각 함수에 minimax_scale방식 정규화 적용

plt.figure(figsize=(20, 5)) # 그림 가로 20인치, 세로 5인치
plt.plot(merge_df['dt'], preprocessing.minmax_scale(merge_df['ncv_kospi']), 'r*:', label='kospi')
plt.plot(merge_df['dt'], preprocessing.minmax_scale(merge_df['ncv_kosdaq']), 'bo--', label='kosdaq')
plt.plot(merge_df['dt'], preprocessing.minmax_scale(merge_df['nv']), 'g^-', label='usd')

# xtick이 항상 10개 출력 되도록
step = len(merge_df) // 10
plt.xticks(merge_df[::step]['dt'])

plt.show()
```

```python
# 각각 함수에 minimax_scale방식 정규화 적용

plt.figure(figsize=(20, 5)) # 그림 가로 20인치, 세로 5인치
plt.plot(merge_df['dt'], preprocessing.minmax_scale(merge_df['ncv_kospi']), 'r*:', label='kospi')
plt.plot(merge_df['dt'], preprocessing.minmax_scale(merge_df['ncv_kosdaq']), 'bo--', label='kosdaq')
plt.plot(merge_df['dt'], preprocessing.minmax_scale(merge_df['nv']), 'g^-', label='usd')

# 라벨 데이터 추가
plt.legend(loc=0)

# xtick이 항상 10개 출력 되도록
step = len(merge_df) // 10
plt.xticks(merge_df[::step]['dt'])

plt.show()
```

```python
def forecast(lat, lng):
    url = "https://api.darksky.net/forecast/{}/{},{}".format(KEY, lat, lng)
    response = requests.get(url)
    data = response.json()
    return data["timezone"], data["hourly"]["summary"]
```
