#### 7개 계층

- Physical
    - 0, 1 볼트
    - 변조 방식
- Data Link
    - 공유기와 Client가 어떤 방식으로 쓸건지?
- Network
    - ip.. 목적지에 대한 정보
- Transport
    - tcp: 신뢰도 보장, eg. 전송이 깨지면 완벽한 전송이 될 때까지 계속 보낸다.. 렉이 걸린다면 tcp
    - udp: 깨져도 그 다음 것으로 보낸다, eg. 영상.. 중간에 깨져도 멈추지 않고 그냥 넘어간다.. live영상이 택하는 방식으로 생각하면 될 듯
- Session
    - 여기서부터는 데이터 부분!
    - 정보를 보내는..?
