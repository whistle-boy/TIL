#### 1. 라이브러리 임포트

````python
import requests
from bs4 import BeautifulSoup
# bs4 라이브러리에서 BeautifulSoup 클래스를 가져온다
````

#### 2. 웹페이지 가져오기

````python
url = 'https://news.v.daum.net/v/20210215150130511'
res = requests.get(url)
# res 객체에 HTMl 파일이 특수한 포맷으로 변수에 넣어진다
````

#### 웹 기본 구조 이해하기
- 인터넷이란 기술이 발달하면서 네트워크 하드웨어 장비가 컴퓨터에 내장되기 시작
- 네트워크
    - 내 컴퓨터와 다른 컴퓨터를 연결
    - 정보를 서로 주고 받을 수 있다
    - 각각의 컴퓨터를 지칭하기 위해 컴퓨터 주소인 'IP Address'
- 인터넷
    - IP Address 를 통해서 특정한 데이터를 가져오거나 보낼 수 있다
    - 주소를 외우기 어려우니 '도메인'
- 웹페이지 동작 원리
    - 클라가 웹브라우저 상에서 도메인을 입력
    - 도메인이 IP 주소로 변환
    - HTML 언어로 구성된 파일을 달라고 요청
    - 서버컴퓨터가 HTML 언어로 구성된 파일을 클라에게 전달
    - 웹브라우저가 HTML 파일을 이쁜 형태로 변환해서 화면에 보여준다

````python
res.content
````

#### 3. 웹페이지 파싱하기
- 파싱
    - 분리 & 해석
    - 문자열을 의미있는 token (어휘 분석의 단위) 로 분해하고 이들로 이루어진 parse tree 를 만드는 과정
- 컴파일
    - 번역
    - 프로그래밍 언어로 되어 있는 소스 파일을 바이너리 파일로 바꿔준다
    - 컴퓨터 언어인 0과 1로 바꾸는 작업
- 파싱을 대신해주는 라이브러리 'bs4'
    - html을 바로 파싱할 수 있다

````python
soup = BeautifulSoup(res.content, 'html.parser')
# 첫번째 인자 : 분석할 html 내용
# 두번째 인자 : parser 프로그램 중 하나
# soup 이란 변수 안에 분석된 내용이 들어가게 된다

soup
````

#### 4. 필요한 데이터 추출하기
- soup.find() 함수로 원하는 부분을 지정하면 된다
- 변수.get_text() 함수로 추출한 부분을 가져올 수 있다

````python
soup.find('h3')
````

````python
mydata = soup.find('h3')
mydata.get_text()
# get_text() 메소드로 html 코드를 없애고 원하는 데이터만 가져온다
````

#### HTML 이해하기
- 웹페이지는 HTML, CSS, JS 로 표현된다
- HTML
    - 웹페이지를 만드는 언어
    - 마크업 언어
    - <태그> 로 시작, </태그> 로 끝
- 마크업 언어
    - 문서나 데이터의 구조를 표현하는 언어
    - 일반적으로 데이터를 기술하는 정도로만 사용
- 프로그래밍 언어
    - 컴퓨터 시스템을 구동시키는 소프트웨어를 작성하기 위한 형식언어
- 코드
    - 컴퓨터 언어로 작성된 문서
        - 마크업 언어 & 프로그래밍 언어
    - 사람이 기계한테 명령
    - 만들고자하는 소프트웨어의 설계도
- 소스코드
    - 프로그램을 사람이 읽을 수 있는 프로그래밍 언어로 기술한 텍스트 파일
- 에디터
    - 프로그래밍 언어도 메모장에 작성할 수 있다
    - 좀 더 빠르게 편리하게 코드를 작성하기 위해서 제공하는 tool
    - HTML 도 일종의 코드이기 때문에 에디터들이 많이 존재한다
        - sublime text 3 등등
- 웹페이지는 HTML 기본 구조를 가진다
```
<!doctype html> -> 이 문서는 HTML 언어를 사용
<html>          -> 여기서 부터 HTML 문서 시작
<head>          -> 문서 전체 정보(eg. 제목)
</head>
<body>          -> 문서 실제 내용 시작
</body>
</html>
```
