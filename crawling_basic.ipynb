#### 3. 웹페이지 파싱하기
- 파싱
    - 분리 & 해석
    - 문자열을 의미있는 token (어휘 분석의 단위) 로 분해하고 이들로 이루어진 parse tree 를 만드는 과정
- 컴파일
    - 번역
    - 프로그래밍 언어로 되어 있는 소스 파일을 바이너리 파일로 바꿔준다
    - 컴퓨터 언어인 0과 1로 바꾸는 작업
- 파싱을 대신해주는 라이브러리 'bs4'
    - html을 바로 파싱할 수 있다

````python
soup = BeautifulSoup(res.content, 'html.parser')
# 첫번째 인자 : 분석할 html 내용
# 두번째 인자 : parser 프로그램 중 하나
# soup 이란 변수 안에 분석된 내용이 들어가게 된다

soup
````

#### 4. 필요한 데이터 추출하기

#### 크롤링 패턴 코드

````python
import requests
from bs4 import BeautifulSoup

res = requests.get('https://news.v.daum.net/v/20210221210242153')
# 크롤링할 페이지 주소 넣기

soup = BeautifulSoup(res.content, 'html.parser')

mydata = soup.find('title')
# 필요한 데이터 추출하는 코드 넣기

print(mydata.get_text())
print(mydata.string)
# 추출한 데이터를 변수에 넣은 후, 원하는 프로그래밍
````

#### HTML 이해하기
- 웹페이지는 HTML, CSS, JS 로 표현된다
- HTML
    - 웹페이지를 만드는 언어
    - 마크업 언어
    - <태그> 로 시작, </태그> 로 끝
- 웹페이지는 HTML 기본 구조를 가진다
```
<!doctype html> -> 이 문서는 HTML 언어를 사용
<html>          -> 여기서 부터 HTML 문서 시작
<head>          -> 문서 전체 정보(eg. 제목)
</head>
<body>          -> 문서 실제 내용 시작
</body>
</html>
```


```
<br>                              -> 엔터
<img src='image.jpg' id='whstle'  -> 속성 중 어떤 태그든 쓸 수 있는 속성 'id'
                                     동일한 img 태그가 여러개 있고, 특정 태그                                      
                                     에 이름을 붙이고 싶을 때 사용
```


- thead는 각 행의 title
    - 없어도 상관은 없다
- border은 테이블 속성 중 하나
    - 테이블 선을 나타낸다
    - 0 이면 선이 나타나지 않는다
- tr
    - 타이틀 내 하나의 행
- th
    - 하나의 행 내 하나의 열
- 필요한 태그는 사전처럼 그때그때 찾아보자

#### 그외
- 마크업 언어
    - 문서나 데이터의 구조를 표현하는 언어
    - 일반적으로 데이터를 기술하는 정도로만 사용
- 프로그래밍 언어
    - 컴퓨터 시스템을 구동시키는 소프트웨어를 작성하기 위한 형식언어
- 코드
    - 컴퓨터 언어로 작성된 문서
        - 마크업 언어 & 프로그래밍 언어
    - 사람이 기계한테 명령
    - 만들고자하는 소프트웨어의 설계도
- 소스코드
    - 프로그램을 사람이 읽을 수 있는 프로그래밍 언어로 기술한 텍스트 파일
- 에디터
    - 프로그래밍 언어도 메모장에 작성할 수 있다
    - 좀 더 빠르게 편리하게 코드를 작성하기 위해서 제공하는 tool
    - HTML 도 일종의 코드이기 때문에 에디터들이 많이 존재한다
        - sublime text 3 등등
