```python
# 리스트끼리 더하면 리스트가 합쳐진다
a = [1,2,3,4,5]
b = list(range(1, 11))
c = a+b
print(c)
```

#### insert ()

```python
a.insert(3, 7) # 특정 인덱스 지점(3)에 값(7)을 삽입
print(a)
```

#### pop ()

```python
a.pop() # 맨 뒤에 자료를 제거
print(a)
```

```python
a.pop(3) # 3번 인덱스 자료 제거
print(a)
```

#### remove ()

```python
a.remove(4) # 특정 값을 제거
print(a)
```

#### index ()

```python
print(a.index(3)) # 특정 값을 찾아서 인덱스 번호로 return
```

#### sort ()

```python
a.sort()
print(a)
```
