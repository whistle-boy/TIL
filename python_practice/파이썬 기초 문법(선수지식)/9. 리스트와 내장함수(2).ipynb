#### enumerate ()

```python
a = [23, 12, 36, 53, 19]

# enumerate는 튜플로 return 해준다
for x in enumerate(a):
    print(x)
```

#### 튜플은 변경이 불가

```python
b = (1, 2, 3, 4, 5)
b[0] = 7
```

```python
for x in enumerate(a):
    print(x[0], x[1])
```

```python
x[0], x[1]
```

```python
for index, value in enumerate(a):
    print(index, value)
```

#### All ()
- all은 모두가 참이어야지 참을 return
- 하나라도 거짓이라면 거짓을 return

```python
if all(60>x for x in a):
    print('YES')
else:
    print('NO')
```

```python
if all(50>x for x in a):
    print('YES')
else:
    print('NO')
# 53일 때 거짓이 된다
```

#### Any ()
- 조건이 한 번이라도 참이 있으면 참을 return

```python
if any(15>x for x in a):
    print('YES')
else:
    print('NO')
```

```python
if any(11>x for x in a):
    print('YES')
else:
    print('NO')
```
