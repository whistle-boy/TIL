- 함수 이름이 없어서 다음과 같이 부르기도 한다
    - 익명의 함수
    - 람다 표현식

```python
def plus_one(x):
    return x+1

print(plus_one(1))
```

```python
# 람다 함수 식을 호출하려면 변수에 할당해야 한다
# plus_two는 함수 이름이 아니라 변수명

plus_two = lambda x: x+2
print(plus_two(1))
```

#### map (func, data)
- 인자가 두 개 (함수, 자료)

```python
a = [1, 2, 3]
print(list(map(plus_one, a)))
```

#### 어떤 내장함수 내 인자로 사용할 때 람다함수는 유용하다

```python
print(list(map(lambda x: x+1, a))) # 익명의 함수 표현식
```
