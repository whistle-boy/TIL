```python
msg='It is Time'
print(msg.upper()) # 원래 원본은 바꾸지 않는다
print(msg.lower())
print(msg)
```

```python
tmp=msg.upper()
print(tmp)
print(tmp.find('T')) # find는 처음 찾은 인덱스 번호를 return
```

```python
print(tmp.count('T')) # count는 몇개가 있는지
```

#### 슬라이싱

```python
print(msg[3:5])
```

```python
for i in range(len(msg)):
    print(msg[i], end=' ')
```

```python
for x in msg:
    print(x, end='') # x가 문자 한개한개를 접근한다
```

```python
for x in msg:
    if x.isupper():
        print(x, end=' ')
```

```python
for x in msg:
    if x.islower():
        print(x, end=' ')
```

#### isalpha ()

```python
for x in msg:
    if x.isalpha(): # 알파벳일 때 참, 공백은 뺀다
        print(x, end='')
```

#### ord ()

```python
tmp = 'AZ'
for x in tmp:
    print(ord(x)) # 아스키 넘버를 출력
```

```python
tmp = 'az'
for x in tmp:
    print(ord(x)) # 아스키 넘버를 출력
```

#### chr ()

```python
tmp = 65
print(chr(tmp)) # 정수를 아스키 넘버로 봤을 때, 문자로 출력
```
