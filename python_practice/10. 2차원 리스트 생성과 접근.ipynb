#### 1차원 리스트 생성

```python
a = [0]*3 # 크기가 3인 리스트
print(a)
```

#### 2차원 리스트 생성

```python
a = [[0]*3 for _ in range(3)] # 반복문이 3번 반복하면서 1차원 리스트를 3번 만듦
print(a)
```

- 2차원 리스트는 표에 개념에서 접근하는 것이 좋다
- 행 * 열

```python
a[0][1] = 1
a[1][1] = 2
a[2][1] = 3
print(a)
```

#### 표처럼 보기

```python
for x in a:
    print(x)
```

```python
# 대괄호 없이 찍고 싶을 때
for x in a:
    for y in x:
        print(y, end=' ')
    print()
```
