- 특정작업을 계속 반복할 때, 함수를 작성하면
    - 코드가 간결해진다
    - 유지보수가 쉬워진다
- 함수는 메인스크립트 위에 작성한다
    - 메인스크립트 밑에 있으면 인터프리터가 해당 함수를 인식할 수 없다

```python
def add(a, b):
    c = a + b
    print(c)
```

```python
add(3, 2)
```

```python
def add(a, b):
    c = a + b
    return c
```

```python
# 함수에서 출력하는 것이 아니라 메인스크립트에서 값을 받아서 출력한다
add(3, 2)
```

```python
x = add(3, 2)
print(x)
```

```python
def add(a, b):
    c = a + b
    d = a - b
    return c, d
# 파이썬은 여러 개의 값을 반환할 수도 있다
# 튜플이라는 자료구조로 리턴이 된다
```

```python
print(add(3, 2))
```

#### 소수만 출력하는 함수
- 소수란 1보다 크고 1과 자기자신으로만 나누어지는 수

```python
def isPrime(x):
    for i in range(2, x): # 2부터 x전까지
        if x%i == 0:
            return False
    return True
# return은 함수를 종료하는 기능도 있다
```

```python
isPrime(12)
```

```python
isPrime(13)
```

```python
a = [12, 13, 7, 9, 19]
```

```python
# 함수에서의 x는 지역변수, 여기서 x는 전역변수
for x in a:
    if isPrime(x): # 만약 False를 리턴하면 그냥 지나가버린다
        print(x, end=' ') # True를 리턴하면 참이 되니까 출력
```
