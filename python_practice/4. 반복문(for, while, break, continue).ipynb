```python
for i in range(20, 10, -1):
    print(i)
```

```python
for i in range(1, 11):
    if i%2==0:
        continue
    print(i)
```

- continue를 만나면 밑에 구문들은 그냥 지나가버린다

#### for / else 구문

```python
for i in range(1, 11):
    print(i)
    if i==5:
        break
```

```python
for i in range(1, 11):
    print(i)
    if i==5:
        break
else:
    print(11)
# 정상적인 종료가 아니면 실행이 안된다
```

```python
for i in range(1, 11):
    print(i)
    if i>15:
        break
else:
    print(11)
# 정상적인 종료가 아니면(break) 실행이 안된다
# 정상 종료면 else를 실행한다
```
